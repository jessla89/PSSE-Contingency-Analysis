### Year 3 Topology 4 Contingency Analysis Reporting - Overloading

```python
import pandas as pd
import numpy as np
```

#### Overload Data 

```python
list_flow = []
list_gens = [0,50,100]
list_lsc = ['lls','rls','hls']
list_gen_hydro = [400,500,600] 
for gen in list_gens:
    for lsc in list_lsc:
        for gen_hy in list_gen_hydro:
            file_out = 'savnw_sol_' + str(gen) +'_hy_' +str(gen_hy) +'_' + str(lsc)+'.xlsx'
            data_flow = pd.read_excel(file_out, sheet_name='Branch Flow')
            data_f_nonan = data_flow.dropna(how='all').reset_index(drop=True)
            data_f_nonan['Scenario']= 'Solar = ' + str(gen) + ' MW, ' +'Hydro = ' +str(gen_hy) +' MW, ' + lsc.upper()
            list_flow.append(data_f_nonan)
data_f = pd.concat(list_flow).reset_index(drop=True)
# branches loaded greater than 100% of their rating
bran_flow_load = data_f[data_f['% FLOW']>100].reset_index(drop=True)
print(len(bran_flow_load))
display(bran_flow_load.head())
```

#### Getting the nonconverged contingencies for each scenario

```python
list_sum = []
list_gens = [0, 50, 100]
list_lsc = ['lls','rls','hls']
list_gen_hydro = [400, 500, 600] 
for gen in list_gens:
    for lsc in list_lsc:
        for gen_hy in list_gen_hydro:
            file_out = 'savnw_sol_' + str(gen) +'_hy_' +str(gen_hy) +'_' + str(lsc)+'.xlsx'
            data_flow = pd.read_excel(file_out, sheet_name='Branch Flow')
            data_f_nonan = data_flow.dropna(how='all').reset_index(drop=True)
            data_f_nonan['Scenario']= 'Solar = ' + str(gen) + ' MW, ' +'Hydro = ' +str(gen_hy) +' MW, ' + lsc.upper()
            data_summary = pd.read_excel(file_out, sheet_name='Summary')
            data_summary_nonan = data_summary.dropna(how='all').reset_index(drop=True)
            data_cont = data_summary_nonan.loc[143:151].reset_index(drop=True)
            # extracting all the tested contingencies 
            tested_conts = list(data_cont['ACCC SOLUTION SUMMARY'])
            ans_list = "".join(tested_conts)
            cont_tested = ans_list.split(',')
            cont_test = [s.strip() for s in cont_tested]
            # extracting contingencies for the studied scenario
            scen_cont = data_f_nonan['CONTINGENCY'].unique()
            # blown up scenarios 
            scenario = 'Solar = ' + str(gen) + ' MW, ' + lsc.upper()
            list_blown = ', '.join(list(set(cont_test)-set(scen_cont)))
            print(f'For the studied scenario, {scenario}, the tested contingencies {list_blown} did not converge')
```

Getting blown up count per scenario 

```python
list_sum = []
list_blown = []
list_gens = [0,50,100]
list_lsc = ['lls','rls','hls']
list_gen_hydro = [400,500,600] 
for gen in list_gens:
    for lsc in list_lsc:
        for gen_hy in list_gen_hydro:
            file_out = 'savnw_sol_' + str(gen) +'_hy_' +str(gen_hy) +'_' + str(lsc)+'.xlsx'
            data_flow = pd.read_excel(file_out, sheet_name='Branch Flow')
            data_f_nonan = data_flow.dropna(how='all').reset_index(drop=True)
            data_f_nonan['Scenario']= 'Solar = ' + str(gen) + ' MW, ' +'Hydro = ' +str(gen_hy) +' MW, ' + lsc.upper()
            data_summary = pd.read_excel(file_out, sheet_name='Summary')
            data_summary_nonan = data_summary.dropna(how='all').reset_index(drop=True)
            data_cont = data_summary_nonan.loc[143:151].reset_index(drop=True)
            # extracting all the tested contingencies 
            tested_conts = list(data_cont['ACCC SOLUTION SUMMARY'])
            ans_list = "".join(tested_conts)
            cont_tested = ans_list.split(',')
            cont_test = [s.strip() for s in cont_tested]
            # extracting contingencies for the studied scenario
            scen_cont = data_f_nonan['CONTINGENCY'].unique()
            # blown up scenarios 
            
            scenario = 'Solar = ' + str(gen) + ' MW, ' +'Hydro = ' +str(gen_hy) +' MW, ' + lsc.upper()
            
            blown_cont = list(set(cont_test)-set(scen_cont))
            list_blown.extend(blown_cont)
```

#### Branch loading data wrangling

```python
bran_overload = bran_flow_load.rename(columns={'BRANCH':'Branch',
                                        'CONTINGENCY':'Contingency',
                                        'MVAFLOW': 'MVA Flow',
                                        'AMPFLOW':'AMP Flow',
                                        'RATE RATE1/RATE2':'Rate', 
                                        '% FLOW':'Loading'})
bran_overload['Contingency'] = bran_overload['Contingency'].str.replace('&','\&')
bran_overload['Branch'] = bran_overload['Branch'].str.replace('_','\_')
```

```python
print(len(bran_overload))
display(bran_overload.head())
```

Overload violations

```python
list_scen = list(bran_overload['Scenario'].unique())
branch_overload = bran_overload[['Scenario','Branch','Contingency','MVA Flow','AMP Flow','Rate','Loading']]
```

Unit Fault Contingency loading > 100% and loading > 130%

```python
print('Unit Fault Contingency loading > 100% and loading > 130%')
print('--------------------------------------------------------------------------------------------------------------------------------')
for scen in list_scen:
    print(scen)
    bran_load_1 = branch_overload[branch_overload['Scenario'] == scen]
    bran_load = bran_load_1[['Branch','Contingency','MVA Flow','AMP Flow','Rate','Loading']]
    unit = bran_load[bran_load['Contingency'].str.contains('UNIT')].reset_index(drop=True)
    if len(unit) == 0:
        print(f'For the studied scenario {scen}, no overloading were reported due to a UNIT Fault')
    if(len(unit) != 0):
        unit_100 = unit[unit['Loading']>100]
        if(len(unit_100) != 0):
            list_ans_single = list([[int(s) for s in m.split() if s.isdigit()] for m in unit_100['Branch'].dropna().unique()])
            ans_single = ', '.join([(str(s[0])+'-'+str(s[1])+'('+ str(s[2])+')') for s in list_ans_single])
            print(f'For the studied scenario {scen}, Loading of > 100% reported for branch(es) {ans_single}')
        unit_130 = unit[unit['Loading']>130]
        if(len(unit_130) != 0):
            list_ans_single = list([[int(s) for s in m.split() if s.isdigit()] for m in unit_130['Branch'].dropna().unique()])
            ans_single = ', '.join([(str(s[0])+'-'+str(s[1])+'('+ str(s[2])+')') for s in list_ans_single])
            print(f'For the studied scenario {scen}, Loading of > 130% reported for branch(es) {ans_single}')
    print('--------------------------------------------------------------------------------------------------------------------------------')
```

Bus Fault Contingency loading > 100% and loading > 130%

```python
print('Bus Fault Contingency loading > 100% and loading > 130%')
print('--------------------------------------------------------------------------------------------------------------------------------')
for scen in list_scen:
    print(scen) 
    bran_load_1 = branch_overload[branch_overload['Scenario'] == scen]
    bran_load = bran_load_1[['Branch','Contingency','MVA Flow','AMP Flow','Rate','Loading']]
    bus = bran_load[bran_load['Contingency'].str.contains('BUS')].reset_index(drop=True)
    if len(bus) == 0:
        print(f'For the studied scenario {scen}, no overloading were reported due to a BUS Fault')
    if len(bus) != 0:
        bus_100 = bus[bus['Loading']>100]
        if(len(bus_100) != 0):
            list_ans_single = list([[int(s) for s in m.split() if s.isdigit()] for m in bus_100['Branch'].dropna().unique()])
            ans_single = ', '.join([(str(s[0])+'-'+str(s[1])+'('+ str(s[2])+')') for s in list_ans_single])
            print(f'Loading of > 100% reported for branch(es) {ans_single}')
        bus_130 = bus[bus['Loading']>130]
        if(len(bus_130) != 0):
            list_ans_single = list([[int(s) for s in m.split() if s.isdigit()] for m in bus_130['Branch'].dropna().unique()])
            ans_single = ', '.join([(str(s[0])+'-'+str(s[1])+'('+ str(s[2])+')') for s in list_ans_single])
            print(f'Loading of > 130% reported for branch(es) {ans_single}')
    print('--------------------------------------------------------------------------------------------------------------------------------')
```

Overload Count - Branches 

```python
list_cont_load = list(bran_overload['Branch'].unique())
list_scen_count_cont = []
list_cont = []
list_scen_cont = []
for cont in list_cont_load:
    bran_load_cont = bran_overload[bran_overload['Branch'] == cont]
    
    no_scenload_cont = len(list(bran_load_cont['Scenario']))
    list_scen_count_cont.append(no_scenload_cont)
    list_ans_cont = list(bran_load_cont['Branch'].unique())
    list_ans_bran = list([[int(s) for s in m.split() if s.isdigit()] for m in bran_load_cont['Branch'].dropna().unique()])
    bran_new = [(str(s[0])+'-'+str(s[1])+'('+ str(s[2])+')') for s in list_ans_bran]
    ans_bran = ', '.join(bran_new)
    list_cont.append(ans_bran)
    ans_scen_cont = ', '.join(bran_load_cont['Scenario'].unique())
    list_scen_cont.append(ans_scen_cont)

dict_overload_bran_cont = {
                    'Branch': list_cont,
                    'Overload Count':list_scen_count_cont,
                    'Scenarios': list_scen_cont
                  
                }
summary_count_load_bran_scen= pd.DataFrame(dict_overload_bran_cont, columns=[ 'Branch','Overload Count']).sort_values(by = 'Overload Count', ascending=False).reset_index(drop=True)
summary_count_load_bran_scen
```
